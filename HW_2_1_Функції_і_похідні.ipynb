1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Визначення функції
x = sp.symbols ('x')
f = 10 * x - 5
```

```python
# Обчислення похідної
derivative_f = sp.diff (f, x)
print("Похідна функції: ", derivative_f)
```

```python
# Діапазон значень для x
x = np.linspace(0, 10, 100)

# Обчислення значень функції та її похідної
y = 10 * x - 5
y_derivative = np.full_like(x, 10)

# Побудова графіків
plt.figure(figsize=(10, 6))

# Графік функції
plt.plot(x, y, label="f(x) = 10x - 5", color="blue")

# Графік похідної
plt.plot(x, y_derivative, label="f'(x) = 10", color="red", linestyle="--")

# Додавання підписів
plt.title("Графік функції f(x) та її похідної")
plt.xlabel("Кількість годин (x)")
plt.ylabel("Оцінка")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid(True)

# Відображення графіків
plt.show()
```

Аналіз:
Кількість годин навчання впливає на оцінку лінійно: більше часу на навчання призводить до вищих оцінок.
Швидкість зростання оцінки не змінюється, незалежно від кількості витрачених годин. 

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, ln, diff, lambdify

# Символьна змінна
x = symbols('x')

# Визначення функцій
f1 = sin(x)
f2 = ln(x)

# Обчислення похідних
f1_derivative = diff(f1, x)
f2_derivative = diff(f2, x)

# Перетворення символьних виразів у функції Python
f1_func = lambdify(x, f1, 'numpy')
f2_func = lambdify(x, f2, 'numpy')
f1_derivative_func = lambdify(x, f1_derivative, 'numpy')
f2_derivative_func = lambdify(x, f2_derivative, 'numpy')

# Діапазон значень для графіків
x_values = np.linspace(0.1, 10, 500)  # Починаємо з 0.1, щоб уникнути ln(0)

# Обчислення значень функцій і похідних
y1_values = f1_func(x_values)
y2_values = f2_func(x_values)
y1_derivative_values = f1_derivative_func(x_values)
y2_derivative_values = f2_derivative_func(x_values)

# Побудова графіків
plt.figure(figsize=(12, 6))

# Графіки f1 і її похідної
plt.subplot(1, 2, 1)
plt.plot(x_values, y1_values, label='f(x) = sin(x)', color='blue')
plt.plot(x_values, y1_derivative_values, label="f'(x) = cos(x)", color='red', linestyle='--')
plt.title('Графік f(x) = sin(x) та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Графіки f2 і її похідної
plt.subplot(1, 2, 2)
plt.plot(x_values, y2_values, label='f(x) = ln(x)', color='green')
plt.plot(x_values, y2_derivative_values, label="f'(x) = 1/x", color='orange', linestyle='--')
plt.title('Графік f(x) = ln(x) та її похідної')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, lambdify

# Задана функція
x = symbols('x')
f = x**2 - 5*x + 1

# Обчислення похідної
f_derivative = diff(f, x)

# Перетворення функцій на числові для використання з numpy
f_numeric = lambdify(x, f, 'numpy')
f_derivative_numeric = lambdify(x, f_derivative, 'numpy')

# Створення відрізка [0, 100]
x_values = np.linspace(0, 100, 500)

# Обчислення значень функції та її похідної
y_values = f_numeric(x_values)
y_derivative_values = f_derivative_numeric(x_values)

# Візуалізація
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Функція f(x)', color='blue')
plt.plot(x_values, y_derivative_values, label="Похідна f'(x)", color='red', linestyle='--')

plt.title('Функція та її похідна')
plt.xlabel('x')
plt.ylabel('Значення')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```
